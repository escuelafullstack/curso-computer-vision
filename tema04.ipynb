```
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.image as mpimg
import sklearn
```

Inicializar arrays numpy

```
lista = [1, 2, 3, 4, 5]
arr = np.array(lista)
print("Data:", arr)

id = np.argmax(arr)
print("id max value: ", id, "\nmax value: ", arr[id])
```

```
arr_float32 = np.asarray(arr, np.float32)
print("Data: ", arr_float32, "\nTipo: ", arr_float32.dtype)
```

```
arr_zeros = np.zeros((5, 2))

print("Data: ")
print(arr_zeros)
print("Shape: ", arr_zeros.shape)

arr_zeros_T = arr_zeros.T
print(arr_zeros_T)
print("Shape arr_T: ", arr_zeros_T.shape)
```

```
arr_zeros_reshape = arr_zeros.reshape((1, 10))
print(arr_zeros_reshape)
```

Propiedad Vectorizacion

```
arr_std =  arr / 5
arr_norm = (arr - arr.mean()) / (arr.max() - arr.min())

print("Array estandarizado: ", arr_std)
print("Array normalizado: ", arr_norm)
```

Categorical Data

```
from sklearn.preprocessing import LabelBinarizer

arr_obj = ['car', 'person', 'cat', 'person']
bin_obj = LabelBinarizer()
arr_encoded = bin_obj.fit_transform(arr_obj)
arr_encoded
```
